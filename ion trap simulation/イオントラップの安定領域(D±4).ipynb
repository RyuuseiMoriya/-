```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

#D4

#----


#--_x--

beta_x = 0.0
q = np.linspace(-2.0, 2.0, 400)#q_x
a_xlist = []
a_zlist = []
q_zlist = []

def f(a_x, q_x, beta_x):#
    
     S_1 = (a_x - (2 + beta_x)**2)*(a_x - (4 + beta_x)**2) - q_x**2#D_4
     S_2 = (a_x - (-2 + beta_x)**2)*(a_x - (-4 + beta_x)**2) - q_x**2#D_-4
     S = (a_x - (4 + beta_x)**2)/ S_1 + (a_x - (-4 + beta_x)**2)/ S_2#
     return  beta_x**2 - a_x + q_x**2 * S#

a_guess = 0.0#

for q_x in q:#q_x

 ax_vals = fsolve(f, a_guess, args= (q_x, beta_x))#

 a_x = ax_vals[0]

 a_xlist.append(a_x)
 a_zlist.append(-2 * a_x)
 q_zlist.append(-2 * q_x)

 a_guess = a_x


#--_x--

beta_x = 1.0
a_xlist1 = []
a_zlist1 = []
q_zlist1 = []


a_guess1 = 0.0#

for q_x1 in q:#q_x

 ax_vals1 = fsolve(f, a_guess1, args= (q_x1, beta_x))#

 a_x1 = ax_vals1[0]

 a_xlist1.append(a_x1)
 a_zlist1.append(-2 * a_x1)
 q_zlist1.append(-2 * q_x1)

 a_guess1 = a_x1


#--z--


#--_z--

beta_z = 0.0
q = np.linspace(-2.0, 2.0, 400)#q_z
a_zlist0 = []
q_zlist0 = []

def f(a_z, q_z, beta_z):#
    
     S_1 = (a_z - (2 + beta_z)**2)*(a_z - (4 + beta_z)**2) - q_z**2#D_4
     S_2 = (a_z - (-2 + beta_z)**2)*(a_z - (-4 + beta_z)**2) - q_z**2#D_-4
     S = (a_z - (4 + beta_z)**2)/ S_1 + (a_z - (-4 + beta_z)**2)/ S_2#
     return  beta_z**2 - a_z + q_z**2 * S#

a_guess = 0.0#

for q_z in q:#q_x

 az_vals = fsolve(f, a_guess, args= (q_z, beta_z))#

 a_z = az_vals[0]

 a_zlist0.append(a_z)
 q_zlist0.append(q_z)

 a_guess = a_z


#--_z--

beta_z = 1.0

a_zlist01 = []
q_zlist01 = []

a_guess = 0.0#

for q_z in q:#q_x

 az_vals = fsolve(f, a_guess, args= (q_z, beta_z))#

 a_z = az_vals[0]

 a_zlist01.append(a_z)
 q_zlist01.append(q_z)

 a_guess = a_z

plt.figure(figsize=(7,5))
plt.plot(q_zlist, a_zlist, color='blue', label=r'$\beta_x = 0$', linewidth = 2)
plt.plot(q_zlist1, a_zlist1, color='red', label=r'$\beta_x = 1$', linewidth = 2)
plt.plot(q_zlist0, a_zlist0, color='green', label=r'$\beta_z = 0$', linewidth = 2)
plt.plot(q_zlist01, a_zlist01, color='orange', label=r'$\beta_z = 1$', linewidth = 2)
plt.xlabel(r'$q_z$')
plt.ylabel(r'$a_z$')
plt.title('Stability diagram at D4')
plt.grid(True)
plt.xlim(0, 1.75)
plt.ylim(-1, 0.5)
plt.legend()
plt.show()
```
